```python
import pandas as pd

breach_data = pd.read_csv('data/processed/sec_company_financials.csv')
col_to_keep = [
    'CIK', 'EntityName', 'Assets', 'Liabilities', 'NetIncome', 'Revenue'
]
breach_data1 = breach_data[col_to_keep]
print("--- Breach Data ---")
print(f'Orig Size: {len(breach_data1)}')
print(f"Missing Entity Names: {breach_data1.isna().sum()['EntityName']}")
print("--- After Entity Name Drop ---")
breach_data2 = breach_data1.dropna(subset=['EntityName'])
print(breach_data2.isna().sum())
print("--------------------")
print(f'Size after null entity name drop: {len(breach_data2)}')
print("Significant null values in date-related columns, liabilities, and revenue")
```

```python
incidents_data = pd.read_csv('data/processed/vcdb_cybersecurity_incidents.csv')
col_to_drop = ['victim_state', 'victim_industry', 'victim_revenue', 'discovery_date', 'reference', 'source_id']
incidents_data2 = incidents_data.drop(col_to_drop, axis=1)
incidents_data2['records_lost'] = incidents_data2['records_lost'].fillna('0')
print("--- Incidents Data ---")
print(f"Orig Size: {len(incidents_data2)}")
print(f"Missing Victim Names: {incidents_data2.isna().sum()['victim_name']}")
print("--- After Victim Name Drop ---")
incidents_data3 = incidents_data2.dropna(subset=['victim_name'])
print(incidents_data3.isna().sum())
print("--------------------")
print(f'Size after null victim name drop: {len(incidents_data3)}')
print("Significant null values in number of records lost, however dataframe.fillna() was utilized to fill missing values as 0")
print("Victim state is insignificant column due to it refering only to states within the US and excluding all other countries.")
```
